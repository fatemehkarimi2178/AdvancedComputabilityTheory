```python
code_list = input('Inter Instructions Code : ').split(' ')
for i in range(0, len(code_list)):
    code_list[i] = int(code_list[i])

for i in code_list:
    
    # <a, <b, c>>
    a, b, c = 0, 0 ,0
    
    # a:
    i = i + 1
    while (i % 2 == 0):
        a = a + 1
        i = i // 2
    # b:
    i = i - 1
    i = i // 2
    i = i + 1
    while (i % 2 == 0):
        b = b + 1
        i = i // 2
    # c:
    i = i - 1
    c = i // 2
    
    var, label, label_prim = '', '', ''
    # variable:
    c = c + 1
    if (c == 1):
        var = 'Y'
    else:
        if (c % 2 == 0):
            var = 'X' + str(c // 2)
        else:
            var = 'Z' + str(c // 2)
    
    # label:
    if (a == 0):
        label = ''
    else:
        if (a % 5 == 0):
            label = '[E' + str(a // 5) + ']'
        if (a % 5 == 1):
            label = '[A' + str(a // 5 + 1) + ']'
        if (a % 5 == 2):
            label = '[B' + str(a // 5 + 1) + ']'
        if (a % 5 == 3):
            label = '[C' + str(a // 5 + 1) + ']'
        if (a % 5 == 4):
            label = '[D' + str(a // 5 + 1) + ']'
    
    # instruction:
    if ( b == 0):
        print (label + ' ' + var + ' <- ' + var )
    elif ( b == 1):
        print (label + ' ' + var + ' <- ' + var + ' + 1')
    elif (b == 2):
        print (label + ' ' + var + ' <- ' + var + ' - 1')
    else:
        b = b - 2
        if (b % 5 == 0):
            label_prim = 'E' + str(b // 5)
        if (b % 5 == 1):
            label_prim = 'A' + str(b // 5 + 1)
        if (b % 5 == 2):
            label_prim = 'B' + str(b // 5 + 1)
        if (b % 5 == 3):
            label_prim = 'C' + str(b // 5 + 1)
        if (b % 5 == 4):
            label_prim = 'D' + str(b // 5 + 1)
        print (label + ' ' + 'IF ' + var + ' != 0 GOTO ' +  label_prim)
    
```
